```python
from path_find import *

class aiPlayer():
    def __init__(self, player):
        self.player = player
        print(self.player.interact("join"))
        status = str.splitlines(player.status("status"))
        self.team = status[2][11:12]
        if (self.team.islower()):
            self.enemies = ['A','B','C','D','E','F','G','H','I','L','M','N','O','P','Q','R','S','T','U','V','Z']
            self.allies = ['a','b','c','d','e','f','g','h','i','l','m','n','o','p','w','r','s','t','u','v','z','x']
        else:
            self.enemies = ['a','b','c','d','e','f','g','h','i','l','m','n','o','p','w','r','s','t','u','v','z']
            self.allies = ['A','B','C','D','E','F','G','H','I','L','M','N','O','P','Q','R','S','T','U','V','Z','X']
            
        self.unshoot = self.allies + ['#', '&', '@', '!', '$']
        self.map = player.process_map()
        self.start = my_position(self.player, self.team, self.map)
        self.flag = flag_position(self.player, self.team, self.map)
        self.finder = PathFinder(self.map, self.start, self.flag)
        self.finder.a_star()
        self.path = self.finder.get_path()
        
    def find_path(self):
        self.map = player.process_map()
        self.start = my_position(player, self.team, self.map)
        self.finder = PathFinder(self.map, self.start, self.flag)
        self.finder.a_star()
        self.path = self.finder.get_path()
        print(self.path)
        self.move()
        
    def move(self):
        cont = 0 
        for x in range(len(self.path) -1):
            current = self.path[x]
            next = self.path[x+1]
            if (cont % 2):
                self.find_path()
#                 self.map = player.process_map()
            
#             # Attack to the left
#             print('N: ' + str(self.map[:current[0],current[1]]))
#             if(any(x in self.map[:current[0],current[1]] for x in self.enemies)
#               and not any(y in self.map[:current[0],current[1]] for y in self.unshoot)):
#                 self.attack('N')

#             # Attack to the right
#             print('S: ' + str(self.map[current[0]+1:,current[1]]))
#             if(any(x in self.map[current[0]:,current[1]] for x in self.enemies)
#                and not any(y in self.map[current[0]:,current[1]] for y in self.unshoot)):
#                 self.attack('S')

#             # Attack left
#             print('E: ' + str(self.map[current[0]][current[1]+1:]))
#             if(any(x in self.map[current[0]][current[1]:] for x in self.enemies)
#                and not any(y in self.map[current[0]][current[1]:] for y in self.unshoot)):
#                 self.attack('E')

#             # attack right
#             print('W: ' + str(self.map[current[0]][:current[1]]))
#             if(any(x in self.map[current[0]][:current[1]] for x in self.enemies)
#                and not any(y in self.map[current[0]][:current[1]] for y in self.unshoot)):
#                 self.attack('W')
                
            

            if current[0] < next[0]:
#                 print(self.player.interact("move","S"))
#                 print(self.player.status('look'))
                self.player.interact("move","S")
                    

            elif current[0] > next[0]: 
                self.player.interact("move","N")
#                 print(self.player.interact("move","N"))
#                 print(self.player.status('look'))

            elif current[1] < next[1]:
                self.player.interact("move","E")
#                 print(self.player.interact("move","E"))
#                 print(self.player.status('look'))

            elif current[1] > next[1]:
                 self.player.interact("move","W")
#                 print(self.player.interact("move","W"))
#                 print(self.player.status('look'))
                
            cont +=1
        
      
    def attack(self, pos):
        print(self.player.interact('shoot', pos))
        print(self.player.status('status'))
        self.find_path()
    
    def print_path(self):
        print("me: " + str(self.team))
        print("my path: " + str(self.path))
        print("my position: " + str(self.start))
        print("my goal: " + str(self.flag))
        print("unshootable items: " + str(self.unshoot))
        
```
